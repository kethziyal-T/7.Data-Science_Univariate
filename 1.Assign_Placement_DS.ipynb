# 1.10 9)Convert the normal distribution to standard normal distribution for
# salary column

# ANS:
         
         Here we covert the normal distribution to standard normal distribution using z -score Z=(X-mean)/(standard)
  
         

```python
def stdNBgraph(dataset):
    
    # Converting to standard normal distribution
    
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    
    values=[i for i in dataset]
    
    z_score=[((j-mean)/std) for j in values]
    
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
    #z_score.std()
```

```python
stdNBgraph(dataset["salary"])
```

# 1.11 10)What is the probability Density Function of the salary range from

# 700000 to 900000?

# ANS:
         
         The Calculated Probability: 
         (P(700,000 between 900,000)=approx 0.00000538 Percentage Chance: 0.00054%.
         
         The Practical Meaning: 
         The probability is practically 0%. Based on the dataset's average(mean) salary approx 2,88,655.
         and standard deviation approx ₹93,457, a salary between 7 Lakhs and 9 Lakhs is an extreme outlier.
         It lies more than 4.4 standard deviations above the mean, making it an exceptionally rare event 
         under a normal curve.

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    
    # generate a sample
    sample = dataset
    
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
    
```

```python
get_pdf_probability(dataset['salary'],700000,900000)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
dataset = pd.read_csv('Placement.csv')

# 2. Filter for placed students (unplaced students have no salary data)
placed_students = dataset[dataset['status'].str.lower() == 'placed'].copy()

# 3. Set up the visualization style
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# 4. Create a grouped boxplot
sns.boxplot(
    data=placed_students, 
    x='specialisation', 
    y='salary', 
    hue='workex', 
    palette='Set2'
)

# 5. Add clear labels and title
plt.title('Salary Distribution by Specialisation and Work Experience', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('MBA Specialisation', fontsize=12)
plt.ylabel('Salary Offered (Rs.)', fontsize=12)
plt.legend(title='Work Experience', loc='upper right')

# 6. Adjust the y-axis layout so numbers do not get squished
plt.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```
